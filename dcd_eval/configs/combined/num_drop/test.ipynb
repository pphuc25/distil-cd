```python
import os

# Directory containing the bash files
directory = './'  # Change to the path of your directory if it's different

# Filename pattern to match
file_pattern = '.sh'

# Text to find and replace
find_text = 'gsm8k_llama2'
replace_text = 'strategy_llama2'

# Iterate through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(file_pattern):
        filepath = os.path.join(directory, filename)

        # Read the contents of the file
        with open(filepath, 'r') as file:
            contents = file.read()

        # Replace the text
        new_contents = contents.replace(find_text, replace_text)

        # Write the modified contents back to the file
        with open(filepath, 'w') as file:
            file.write(new_contents)

        print(f"Updated file: {filename}")

print("All files have been updated.")
```

```python
import os

# Directory containing the files
directory = './'  # Change to the path of your directory if it's different

# Text to find and replace in the file names
find_text = 'csv4'
replace_text = 'csv1'

# Iterate through all files in the directory
for filename in os.listdir(directory):
    if find_text in filename and filename.endswith(".sh"):
        # Create the new file name
        new_filename = filename.replace(find_text, replace_text)
        old_filepath = os.path.join(directory, filename)
        new_filepath = os.path.join(directory, new_filename)

        # Rename the file
        os.rename(old_filepath, new_filepath)

        print(f"Renamed file: {filename} to {new_filename}")

print("All relevant files have been renamed.")
```

```python
import os
import re

# Directory containing the bash files
directory = './'  # Update this with the correct directory path

# Iterate through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".sh") and 'strategyqa_csv' in filename:
        # Construct the expected pattern from the filename
        expected_dropout_match = re.search(r'drop(\d+)', filename)
        expected_beta_match = re.search(r'beta(\d+)', filename)
        if expected_dropout_match and expected_beta_match:
            expected_dropout = int(expected_dropout_match.group(1)) / 10
            expected_beta = int(expected_beta_match.group(1)) / 10

            # Read the contents of the file
            with open(os.path.join(directory, filename), 'r') as file:
                contents = file.read()

                # Extract actual dropout_num and beta_coef values
                actual_dropout_match = re.search(r'--dropout_num (\d+\.\d+)', contents)
                actual_beta_match = re.search(r'--beta_coef (\d+\.\d+)', contents)

                if actual_dropout_match and actual_beta_match:
                    actual_dropout = float(actual_dropout_match.group(1))
                    actual_beta = float(actual_beta_match.group(1))

                    # Check if the values match
                    if actual_dropout != expected_dropout or actual_beta != expected_beta:
                        print(f"Mismatch in file: {filename}")
                else:
                    print(f"Missing parameters in file: {filename}")
        else:
            print(f"Filename pattern not recognized: {filename}")

print("File check complete.")
```
